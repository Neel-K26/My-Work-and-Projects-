# NORMALISATION 

```python
df=pd.DataFrame([
    [25000,200,30.5],
    [10000,300,10.5],
    [9000,100,25.7],
    [40000,300,21.2],
    ],
    columns= ['A','B','C']
)
```

```python
# Display the Data
display(df)
```

```python
# Minimizing Difference between values
 # visulisation
import matplotlib.pyplot as plt
df.plot(kind='bar')
```

# Technique 1; Maximum  Absolute Value
 Rescales each value between -1 to 1
 Divides each observation by its MAS
 use max() and abs() methods


```python
# Creating a Copy of dataset so that we could do numerous possible operations
```

```python
d=df.copy()
d
```

```python
# applying formula of MAV
for col in d.columns:
            d[col]= d[col] / d[col].abs().max()
```

```python
# DAta Normalised
display(d)
```

```python
# now visulaising graph
d.plot(kind='bar')
```

# MIN MAX TECHNIQUE



```python
# SUBTRACTING  min value from  observation and dividing by the range
# Range = Max Obs- Min Obs
# will scale in 0 to 1
```

```python
# Creating another Duplicate
d1=df.copy()
d1
```

```python
for col in d1.columns:
    d1[col] = (d1[col]- d1.min()) / (d1[col].max() - d1[col].min())
    
```

```python
display(d1)
```

```python
d1.plot(kind='bar')
```
