```python
import matplotlib.pyplot as plt
```

```python
# generating a Sine Wave
import math
import numpy as np
x= np.arange(0,math.pi*2,0.05)
y= np.sin(x)
```

```python
plt.xlabel('angle')
plt.ylabel('sine')
plt.plot(x,y)
```

```python
plt.figure(figsize=(16,9), facecolor='red', dpi=100) # figure- aspect ratio
plt.xlabel("X Side")                        # dots-per-inch
plt.ylabel('Y Side')                       
plt.plot(x,y)
```

```python
# Colour and Marker
x=[3,6,12,23,46,67,78,88,90,105]
y=[12,45,56,67,69,72,79,88,94,111]
plt.xlabel('X Side')
plt.ylabel('Y Side')
plt.plot(x,y, color='grey', marker='*')
```

```python
plt.xlabel('X Side')
plt.ylabel('Y Side')
plt.plot(x,y, color='#D56745')  # we can also use color code
```

```python
plt.xlabel('X Side')
plt.ylabel('Y Side')
plt.plot(x,y, color='r', linestyle='-.')    # using linestyle for line graphs
```

```python
# Multiple Graphs
x=[3,6,12,23,46,67,78,88,90,105]
y=[12,45,56,67,69,72,79,88,94,111]
z=[23,34,44,44,47,59,61,69,78,96]
plt.xlabel('X Side')
plt.ylabel('Y Side')
plt.plot(x,y, color='r', linestyle='-.')
plt.plot(x,z , color='g', marker ='.')
```

# Printing same graph differently 

```python
plt.subplot(2,1,1,facecolor='lightblue')   #(rows,columns,box no)
plt.plot(x,y,color='r',linestyle='-.')
plt.subplot(2,1,2,facecolor='lightgray')
plt.plot(x,z,color='g', marker='.')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,color='r',linestyle='-.')
plt.subplot(2,2,3)
plt.plot(x,z,color='g',marker='.')
plt.subplot(2,2,4)
plt.plot(z,y, color='g',marker='.')
```

```python
# Lim function
x=[3,6,12,23,46,67,78,88,90,105]
y=[12,45,56,67,69,72,79,88,94,111]
plt.xlabel('X Side')
plt.ylabel('Y Side')
plt.xlim(0,150)
plt.ylim(0,150)
plt.plot(x,y)
```

```python
# Ticks Function
plt.xlabel('X Side')
plt.ylabel('Y Side')
plt.xticks(range(0,125,10)) # 0-125 incerement by 10
plt.yticks(range(0,125,10))
plt.grid()  # grid functions
plt.plot(x,y) 
```

```python
# Labelling  and Legend
x=[3,6,12,23,46,67,78,88,90,105]
y=[12,45,56,67,69,72,79,88,94,111]
z=[23,34,44,44,47,59,61,69,78,96]
plt.xlabel('X Side')
plt.ylabel('Y Side')
plt.plot(x,y, color='r', linestyle='-.', label='2022')
plt.plot(x,z , color='g', marker ='.',label='2023')
plt.legend()
```

```python
# Scatter Plot
plt.xlabel('X Side')
plt.ylabel('Y Side')
plt.scatter(x,y, color='red', marker='*')
```

# 3d Plot

```python
from mpl_toolkits import mplot3d
fig= plt.figure() # for presenting figure
ax=plt.axes(projection='3d') # axis of projection as 3d
z=np.linspace(0,1,100) # linearly- spaced 100 elements
x=[3,6,12,23,46,67,78,88,90,105]
y=[12,45,56,67,69,72,79,88,94,111]
z=[23,34,44,44,47,59,61,69,78,96]
ax.plot3D(x,y,z,'gray')  # ax.(type of plot)
ax.set_title('3d Line Plot') # Title of the graph
```

```python
fig= plt.figure() # for presenting figure
ax=plt.axes(projection='3d') # axis of projection as 3d
x=np.array([3,6,12,23,46,67,78,88,90,105])
y=np.array([12,45,56,67,69,72,79,88,94,111])
z=np.array([23,34,44,44,47,59,61,69,78,96])
c=x+y
ax.scatter(x,y,z,c=c)  # ax.(type of plot)
ax.set_title('3d Scatter Plot') # Title of the graph
```

# Bar Graph

```python
x=[5,8,10]
y=[12,16,6]

x2=[6,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g',align='center')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```
