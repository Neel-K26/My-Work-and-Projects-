```python
#Distplot
from matplotlib import pyplot as plt
import seaborn as sns
df=sns.load_dataset('iris')
sns.distplot(df['petal_length'],hist=False)
plt.show()
```

```python
#KDE Parameter
df=sns.load_dataset('iris')
sns.distplot(df['petal_length'])
plt.show
```

```python
#Joint Function

df=sns.load_dataset('iris')
sns.jointplot(x='petal_length',y='petal_width',data=df)
plt.show
```

```python
df=sns.load_dataset('iris')
sns.jointplot(x='petal_length',y='petal_width',data=df,kind='hex')
plt.show
```

```python
df=sns.load_dataset('iris')
sns.jointplot(x='petal_length',y='petal_width',
              data=df,kind='kde')
```

```python
#Pair Plot
df=sns.load_dataset('iris')
sns.set_style('ticks')
sns.pairplot(df,hue='species',diag_kind='kde',kind='scatter',palette='husl')
```

```python
#StripPlot Function
df=sns.load_dataset('iris')
sns.stripplot(x='species',y='petal_length',data=df,)
```

```python
#Jitter Function
df=sns.load_dataset('iris')
sns.stripplot(x='species',y='petal_length',data=df,jitter=False)
```

```python
sns.swarmplot(x='species',y='petal_length',data=df)
```

```python
#Box-Plot
df=sns.load_dataset('iris')
sns.boxplot(x='species',y='petal_length',data=df)
```

```python
#Factor-Plot
df=sns.load_dataset('exercise')
sns.factorplot(x='time',y='pulse',hue='kind',data=df)
```

```python
#Factor Plot with Kind
df=sns.load_dataset('exercise')
sns.factorplot(x='time',y='pulse',hue='kind',kind='violin',data=df)
```

```python
df=sns.load_dataset('titanic')
sns.factorplot('alive',col='deck',col_wrap=3,
               data=df[df.deck.notnull()],kind='count')
```

```python
#RegPLot & LmPlot
df=sns.load_dataset('tips')
sns.regplot(x='total_bill',y='tip',data=df)
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
df=sns.load_dataset('tips')
sns.lmplot(x='size',y='tip',data=df)
```

```python
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
g=sns.FacetGrid(df,col='time')
g.map(plt.hist,'tip')
```

```python
#PairGrid
df=sns.load_dataset('iris')
g=sns.PairGrid(df)
g.map(plt.scatter)
```

```python
#PairGrid-Hist&Scatter
df=sns.load_dataset('iris')
g=sns.PairGrid(df)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

```python
df=sns.load_dataset('iris')
g=sns.PairGrid(df)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot,cmap='Blues_d')
g.map_diag(sns.kdeplot,lw=3,legend=False)
```
