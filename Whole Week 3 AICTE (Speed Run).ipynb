# Control structures and Functions

```python
# If, elif, Else
# creating new columns
cars.insert(10,'Price_Class','') #(position,column name,''[blank values])
```

```python
for i  in range(0,len(cars['Price']),1):
    if(cars['Price'][i]<8450):
        cars['Price_Class'][i]='Low'
    elif((cars['Price'][i]>11950)):
        cars['Price_Class'][i]='High'
    else: cars["Price_Class"][i]='Medium'
```

```python
i=0
while i<len(cars['Price']):
    if(cars['Price'][i]<8450):
        cars['Price_Class'][i]='Low'
    elif((cars['Price'][i]>11950)):
        cars['Price_Class'][i]='High'
    else: cars["Price_Class"][i]='Medium'
    i=i+1
```

```python
# Count of unique values(series)
cars["Price_Class"].value_counts()
```

```python
# Functions in Python
# def function _name(paramaters): statements
# 1] Converting age=> months--> years

#new column
cars.insert(11,'Age_Convert',0)
```

```python
# new function c_convert returns 1 value
def c_convert(val):
    val_converted=val/12
    return val_converted
```

```python
cars["Age_Convert"]=c_convert(cars['Age'])
cars['Age_Convert']=round(cars['Age_Convert'],1)
```

```python
# Function with multiple inputs and outputs
# Converting Age to months=> years
# Km to run per month
def c_convert(val1,val2): # 2 inputs
    val_converted=val1/12
    ratio=val2/val1
    return[val_converted,ratio] # 2 outputs
```

# Exploratory Data Analysis


```python
data=pd.read_csv('Toyota.csv',index_col=0,na_values=['??','????'])
```

```python
cars=data.copy()
```

```python
# Frequency Tables
# crosstab()==> to count frequenc of all elements
pd.crosstab(index=cars['FuelType'],columns='count',dropna=True)
```

```python
cars.size
```

```python
# Two-Way Tables
pd.crosstab(index=cars['Automatic'],columns=cars['FuelType'],dropna=True)
# Automatic cars for each fueltype cars
# rows[index]=> automatic and columns[fueltype]
```

```python
# Two-Way Tables==> joint probablity[2 independent events at same time]
pd.crosstab(index=cars['Automatic'],columns=cars['FuelType'],normalize=True,dropna=True)
# normalize is converting numbers to proportions
```

```python
# Marginal Probablity==> prob of occurence of single event
pd.crosstab(index=cars['Automatic'],columns=cars['FuelType'],margins=True,normalize=True,dropna=True)
```

```python
# Conditional Probablity
# Prob of event A given that event B has already occured (Prob of getting an event)
pd.crosstab(index=cars['Automatic'],columns=cars['FuelType'],margins=True,normalize='index',dropna=True)
```

```python
pd.crosstab(index=cars['Automatic'],columns=cars['FuelType'],margins=True,normalize='columns',dropna=True)
```
